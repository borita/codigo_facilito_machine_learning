```python
import itertools
import pandas as pd
from sklearn.linear_model import LinearRegression
from sklearn.metrics import mean_squared_error, r2_score

# Asumiendo que df ya está cargado y limpio

# Seleccionamos variables numéricas y quitamos la variable objetivo
df_num = df.select_dtypes(include=['number'])
X_all = df_num.drop(columns=['Calificacion'])
y = df_num['Calificacion']

# Lista de nombres de variables predictoras
variables = X_all.columns.tolist()

resultados = []

# Probar todas las combinaciones de 2 variables
for combo in itertools.combinations(variables, 2):
    X_combo = X_all[list(combo)]

    model = LinearRegression()
    model.fit(X_combo, y)
    y_pred = model.predict(X_combo)

    mse = mean_squared_error(y, y_pred)
    r2 = r2_score(y, y_pred)

    resultados.append({
        'Variables': combo,
        'MSE': mse,
        'R2': r2
    })

# Mostrar resultados ordenados por mejor R2
resultados_df = pd.DataFrame(resultados).sort_values(by='R2', ascending=False).reset_index(drop=True)
print(resultados_df)
```

```python
import numpy as np
import pandas as pd

np.random.seed(123)
n = 500

# IDs fijos
ids = [f"S{i:03d}" for i in range(1, n+1)]

# Datos base fijos
genero = np.random.choice(['M', 'F'], size=n)
edad = np.random.randint(10, 19, size=n)
grado = np.random.choice([9, 10, 11, 12], size=n)
becado = np.random.choice([0, 1], size=n, p=[0.8, 0.2])

# Matriz covarianza para horas
cov_matrix = np.array([
    [1.0, 0.8, 0.75],
    [0.8, 1.0, 0.7],
    [0.75, 0.7, 1.0]
])
mean_hours = [5, 4, 3]

def generar_horas(n, becado):
    base = np.random.multivariate_normal(mean_hours, cov_matrix, size=n)
    base += becado.reshape(-1, 1) * np.array([2, 1.5, 1.2])
    base += np.random.normal(0, 0.2, base.shape)
    base = np.clip(base, 0, None)
    return base

def generar_calificacion(horas):
    calif = 10 + 9*horas[:, 0] + 6*horas[:, 1] + 4*horas[:, 2] + np.random.normal(0, 3, len(horas))
    return np.clip(calif, 0, 100)

df_base = pd.DataFrame({
    'ID': ids,
    'Genero': genero,
    'Edad': edad,
    'Grado': grado,
    'Becado': becado
})

def generar_mes(df_base):
    horas = generar_horas(len(df_base), df_base['Becado'].values)
    calif = generar_calificacion(horas)
    df = df_base.copy()
    df['Horas de Estudio'] = horas[:, 0]
    df['Horas de Tarea'] = horas[:, 1]
    df['Horas de Tutoring'] = horas[:, 2]
    df['Calificacion'] = calif
    return df

df_mes_1 = generar_mes(df_base)
df_mes_2 = generar_mes(df_base)
df_mes_3 = generar_mes(df_base)

df_mes_4 = generar_mes(df_base)

# Insertar nulos 5%
for col in ['Horas de Estudio', 'Horas de Tarea', 'Horas de Tutoring', 'Calificacion']:
    idx_null = np.random.choice(df_mes_4.index, size=int(0.05*n), replace=False)
    df_mes_4.loc[idx_null, col] = np.nan

# Insertar outliers 3%
for col in ['Horas de Estudio', 'Horas de Tarea', 'Horas de Tutoring', 'Calificacion']:
    idx_out = np.random.choice(df_mes_4.index, size=int(0.03*n), replace=False)
    df_mes_4.loc[idx_out, col] = df_mes_4[col].mean() + 5*df_mes_4[col].std()

# Guardar archivos .txt
df_mes_1.to_csv('mes_1.txt', index=False)
df_mes_2.to_csv('mes_2.txt', index=False)
df_mes_3.to_csv('mes_3.txt', index=False)
df_mes_4.to_csv('mes_4.txt', index=False)

print("Archivos mes_1.txt, mes_2.txt, mes_3.txt, mes_4.txt generados con éxito.")

from google.colab import files
files.download('mes_1.txt')
files.download('mes_2.txt')
files.download('mes_3.txt')
files.download('mes_4.txt')
```
